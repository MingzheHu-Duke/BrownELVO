```python
import chardet
import pandas as pd
```

```python
# Read in the elvo keys
with open('/content/train_elvo_detection_kfold_with_original_accessions.csv', 'rb') as f:
    result = chardet.detect(f.read())  # or readline if the file is large


df_keys = pd.read_csv('/content/train_elvo_detection_kfold_with_original_accessions.csv', encoding=result['encoding'])
df_keys.head(10)
```

```python
# Read in the clinical info
df_info = pd.read_excel('/content/BROWN急慢性判断ELVO_official.xlsx')
df_info.head(10)
```

```python
# Find the maps between the Sham Accession number and the accession number
df_maps = df_keys[["accession"]]
# Peek
df_maps.head()
```

```python
# Left Merge to extract the info
df_merged = df_maps.merge(df_info, left_on="accession", right_on="Accession Number")
```

```python
print("Len of extracted clinical info:" +str(len(df_merged)))
print("Len of keys:" +str(len(df_keys)))
print("Len of original Brown exel:" +str(len(df_info)))
```

```python
# Save the data frameinto csv file
df_merged.to_csv ('extracted_clinical_info.csv', index = False, header=True)
```
